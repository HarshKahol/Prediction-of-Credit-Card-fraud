```python
def credit(df):
    o = pd.DataFrame()
    for i in df.columns:
        mean_vale = np.mean(df[i])
        std_vale = np.std(df[i])
        z = (df[i]- mean_vale)/std_vale
        o[i] =df[i][(z>3)|(z<-3)]
    return o
        
```

```python
outliers = credit(df)
```

```python
outliers
```

```python
def credit_without_outlier(df):
    cleaned = df.copy()
    for i in df.columns:
        m = np.mean(df[i])
        s = np.std(df[i])
        z = (df[i]- m)/s
        cleaned = cleaned[(z<3) & (z>-3)]
    return cleaned
```

```python
cleaned = credit_without_outlier(df)
```
